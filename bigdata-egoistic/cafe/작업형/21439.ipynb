### Q2. 각 일자의 요일을 표기하고 ('Monday' ~'Sunday') 'day_name'컬럼을 추가하고 이를 이용하여 각 요일별 이용 횟수의 총합을 데이터 프레임으로 출력하라

```python
df['대여일자'] = pd.to_datetime(df['대여일자'])
df['day_name'] = df['대여일자'].dt.day_name()
print(df['day_name'].value_counts().to_frame())
```

### Q3. 각 요일별 대여소 이용 횟수를 데이터프레임으로 출력하라

```python
df.info()
```

```python
df1 = df.loc[:100, :]
```

```python
df1.head(100)
```

```python
result = df1.groupby(['day_name','대여소번호']).size().to_frame('size').sort_values(['day_name','size'], ascending=False).reset_index()
```

```python
result
```

```python
answer1 = result.drop_duplicates(['day_name'],keep='first').reset_index(drop=True)
answer2 = result.drop_duplicates(['day_name', '대여소번호'],keep='first').reset_index(drop=True)
answer3 = result.drop_duplicates(['day_name', '대여소번호', 'size'],keep='first').reset_index(drop=True)
```
